# Exploratory Data Analysis (EDA) on Iris Dataset:

## libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

## Load Dataset:

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
df.head()
```

## Dataset Information:

```python
df.info()
```

## Statistical Summary:

```python
df.describe()
```

## Check Missing Values:

```python
df.isnull().sum()
```

## Histograms of Features:

```python
df.hist()
plt.show()
```

```python
df.describe()
```

## Pairplot:

```python
sns.pairplot(df, hue="species")
plt.show()
```

## Correlation Heatmap:

```python
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Scatter Plot Example:

```python
plt.figure()
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.title("Scatter Plot of First Two Features")
plt.show()
```



## Conclusion

- The Iris dataset contains well-separated classes.
- Petal length and petal width are the most important features.
- Data visualization helps in understanding feature relationships clearly.
